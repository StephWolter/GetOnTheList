```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
from pprint import pprint as pp
from datetime import datetime
```

```python
from webdriver_manager.chrome import ChromeDriverManager
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

### For social media


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://www.searchenginejournal.com/social-media/biggest-social-media-sites/#close"
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
media_soup = soup(html, 'html.parser')
media_soup
```

```python
# Extract all rows of data
media_table = media_soup.find('div', class_='scrl-table')
print(media_table)
```

```python
media_info = []
table_rows = media_table.find_all('div', class_ = 'scrl-table')

# Loop through the scraped data to create a list of rows
for row in table_rows:
    td = row.find_all('td')
#     row = [col.text for col in td]
    media_info.append(row)

media_info
```

```python
media_info = []
table_rows = media_table.find_all('tr', class_ = 'data-row')

# Loop through the scraped data to create a list of rows
for row in table_rows:
    td = row.find_all('td')
    row = [col.text for col in td]
    media_info.append(row)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
table_headers = media_table.find_all('th')
headers = []
for header in table_headers:
    headers.append(header.text)

media_df = pd.DataFrame(media_info, columns = headers)
```

```python
# Confirm DataFrame was created successfully
media_df
```
